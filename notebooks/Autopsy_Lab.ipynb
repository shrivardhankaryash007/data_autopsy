```python
# Pass-1 demo (toy CSV)
from pathlib import Path

import pandas as pd

from autopsy.store import MeasurementStore

store = MeasurementStore(root=Path('.autopsy_cache_demo'))

toy_path = Path('tmp_pass1_demo.csv')

df = pd.DataFrame({
    'timestamp': [0, 0.5, 1, 1.5, 2, 2.5],
    'signal_a': [1, 1, 2, 3, 4, 5],
    'signal_b': [0, 0, 0, 0, 20, 20],
})
df.to_csv(toy_path, index=False)

ref = store.add(toy_path, label='toy_pass1')
overview_cfg = {
    'signals': ['signal_a', 'signal_b'],
    'hz': 1.0,
    'agg': ['min', 'mean', 'max'],
    'time_col': 'timestamp',
}
store.build_overview(ref.id, **overview_cfg)

pass1_cfg = {
    'missing_rate': 0.1,
    'flatline_eps': 0.01,
    'flatline_min_run': 2,
    'spike_mad_z': 3.0,
    'top_k_windows': 3,
    'top_n_signals': 2,
}

result = store.run_autopsy_pass1(ref.id, overview_cfg, pass1_cfg)
print(result['executive_summary_path'])
print(Path(result['executive_summary_path']).read_text())
```
